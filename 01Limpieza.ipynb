# Limpieza avanzada del texto
Este notebook aplica una limpieza robusta al texto para eliminar símbolos extraños, emojis, y otros caracteres no deseados antes de la vectorización.

```python
import pandas as pd
import re

# Cargar el dataset
df = pd.read_csv("Tweets.csv")  # Cambia si es necesario
df.head()
```

```python
# Función de limpieza
def limpiar_texto(texto):
    if pd.isna(texto):
        return ""
    texto = texto.lower()
    texto = re.sub(r"http\S+", "", texto)
    texto = re.sub(r"[^a-zA-Z0-9\s]", "", texto)
    texto = re.sub(r"\s+", " ", texto).strip()
    return texto
```

```python
# Aplicar limpieza a la columna de texto
df["text_limpio"] = df["text"].apply(limpiar_texto)
df[["text", "text_limpio"]].head()
```

```python
# Guardar resultado
df.to_csv("01tweetsLimpios.csv", index=False)
print("✅ Limpieza completa. Archivo guardado como 01tweetsLimpios.csv")
df.head(10)
```
